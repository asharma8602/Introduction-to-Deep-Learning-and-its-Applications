```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount("/content/gdrive")
df = pd.read_csv('/content/gdrive/My Drive/House_prediction.csv')
```

```python
df_prices = df.groupby("city").agg(np.mean)
df_prices.plot.bar(y='rent amount (R$)')
df_prices1 = df.groupby("rooms").agg(np.mean)
df_prices1.plot.bar(y='rent amount (R$)')
df_prices2 = df.groupby("floor").agg(np.mean)
df_prices2.plot.bar(y='rent amount (R$)')
```

```python
df_room = df.groupby("rooms").agg("mean")
df_room.plot.bar(y='hoa (R$)')
df_room.plot.bar(y='property tax (R$)')
df_room.plot.bar(y='fire insurance (R$)')
```

```python
df_furn = df.groupby("furniture").agg("mean")
df_furn.plot.bar(y='hoa (R$)')
df_furn.plot.bar(y='property tax (R$)')
df_furn.plot.bar(y='fire insurance (R$)')
```

```python
df_furn1 = df.groupby("bathroom").agg("mean")
df_furn1.plot.bar( y='hoa (R$)')
df_furn1.plot.bar( y='property tax (R$)')
df_furn1.plot.bar( y='fire insurance (R$)')
```
