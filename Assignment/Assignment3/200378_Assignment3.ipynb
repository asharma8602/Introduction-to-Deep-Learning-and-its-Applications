```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, LSTM 

```

```python
mnist = tf.keras.datasets.mnist
(xtrain, ytrain), (xtest, ytest) = mnist.load_data()

xtrain = xtrain/255.0
xtest = xtest/255.0    #normalising the data 
rnn = Sequential()


#defining the model 
#print(xtrain[0])
print(ytrain[0])


```

```python
rnn.add(LSTM(128, input_shape = (xtrain.shape[1:]), activation = 'relu' , return_sequences =True))
rnn.add(Dropout(0.2))



rnn.add(LSTM(128, activation = 'relu'))
rnn.add(Dropout(0.2))

rnn.add(Dense(32, activation = 'relu'))
rnn.add(Dropout(0.2))

rnn.add(Dense(10, activation = 'softmax'))

#building the architecture, specifying input size and activation functions 


rnn.compile( optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

rnn.fit(xtrain, ytrain, epochs = 2, validation_data = (xtest,ytest)) 
```

```python
predicted = rnn.predict(xtest[0])
print(predicted)
```
