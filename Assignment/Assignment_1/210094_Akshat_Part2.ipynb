```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data=pd.read_csv('House_prediction.csv')
col=['area']
#pivot_data=data.pivot(index='city',columns=col,values=)
# print(data)
mean_area=data.groupby('city')['area'].mean()
mean_rooms=data.groupby('city')['rooms'].mean()
mean_bathroom=data.groupby('city')['bathroom'].mean()
mean_parking_spaces=data.groupby('city')['parking spaces'].mean()
#mean_floor=data.groupby('city')['floor'].mean()
mean_hoa=data.groupby('city')['hoa (R$)'].mean()
mean_rent=data.groupby('city')['rent amount (R$)'].mean()
mean_property=data.groupby('city')['property tax (R$)'].mean()
mean_fire=data.groupby('city')['fire insurance (R$)'].mean()
mean_total=data.groupby('city')['total (R$)'].mean()

fig1=plt.figure()
fig1=plt.plot(mean_area)
plt.title('AREA')
plt.show(fig1)
```

```python
fig9=plt.figure()
fig9=plt.plot(mean_total)
plt.title('TOTAL')
plt.show(fig9)
```

From the above graphs we can see Campinas has low total as well as it also have quite good area, rooms, bathrooms and parking space. Porto Alegre has lowest total but other features are very much compromised.
