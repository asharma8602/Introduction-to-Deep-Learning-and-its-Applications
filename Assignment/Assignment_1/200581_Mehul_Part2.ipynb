```python
import pandas as pd
import matplotlib as plt
```

```python
df=pd.read_csv(r"C:\Users\user\Downloads\House_prediction.csv")
```

```python
cities =df.city.unique()
```

```python
print(cities)
```

```python
df.groupby('city').mean().plot(kind='bar', figsize=(12,10))
```

```python
df.plot(x='hoa (R$)' , y='rent amount (R$)', style='o')
# House association tax seems to be almost constant with increasing rent amount

df.plot(x='hoa (R$)' , y='area', style='o')
# Same with area

df.plot(x='hoa (R$)' , y='total (R$)', style='o')
# Same with total amount
```

```python
df.plot(x='property tax (R$)' , y='rent amount (R$)', style='o', color='green')
# Property tax seems to be constant with increasing rent amount

df.plot(x='property tax (R$)' , y='area', style='o', color='green')
# Same with area

df.plot(x='property tax (R$)' , y='total (R$)', style='o', color='green')
# Same with area
```

```python
df.plot(x='fire insurance (R$)' , y='rent amount (R$)', style='o', color='red')
# Fire insurance seemd to be increasing with increasing rent amount 

df.plot(x='fire insurance (R$)' , y='area', style='o', color='red')
# Area seems constant with increasing fire insurance

# df.plot(x='fire insurance (R$)' , y='bathroom', style='o', color='red')

df.plot(x='fire insurance (R$)' , y='total (R$)', style='o', color='red')
# Total amount seems constant with increasing fire insurance
```
