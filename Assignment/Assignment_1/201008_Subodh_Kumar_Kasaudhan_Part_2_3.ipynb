Importing the dependencies and the module required

```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

To read CSV file and plot the average values city wise

```python
url = 'https://raw.githubusercontent.com/shivam-iitk/Introduction-to-Deep-Learning-and-its-Applications/main/Assignment/Assignment_1/House_prediction.csv'
M = pd.read_csv(url)
# used github to get files
# Dataset is now stored in a Pandas Dataframe
temp = list(M.head())
data = M.groupby(["city"])[temp[2:]]
data = data.mean()
print(len(data.head(5)))
for i in range(len(data.head(8))):
    plt.figure(figsize=(21, 9))
    x = data.index.to_list()
    y = data.loc[:,list(data.head(0))[i]]
    plt.plot(x,y)
    plt.ylabel(list(data.head(0))[i])
    plt.show()
data
```

```python
corr=M.corr()
plt.figure(figsize=(15,15))
sns.heatmap(corr,cbar=True,square=True,fmt='.2f',annot=True, annot_kws={'size':15}, cmap='plasma')
```
