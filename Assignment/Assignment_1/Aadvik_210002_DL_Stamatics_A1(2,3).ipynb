```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


A = pd.read_csv('House_prediction.csv')


ax1 = A.groupby("city").agg(np.mean)
ax1.plot.barh(y='rent amount (R$)' , color='pink')

ax2 = A.groupby("city").agg(np.mean)
ax2.plot.barh(y='rooms' , color='darkred')

ax3 = A.groupby("city").agg(np.mean)
ax3.plot.barh(y='area' , color='purple')

ax4 = A.groupby("city").agg(np.mean)
ax4.plot.barh(y='parking spaces' , color='darkblue')


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


A = pd.read_csv('House_prediction.csv')


ax1 = A.groupby("rooms").agg(np.mean)
ax1.plot.barh( y = 'hoa (R$)' , color='pink')

ax1.plot.barh(y= 'property tax (R$)' , color='darkred')

ax1.plot.barh(y= 'fire insurance (R$)' , color='purple')

ax1.plot.barh(y= 'total (R$)' , color='darkblue')
```
