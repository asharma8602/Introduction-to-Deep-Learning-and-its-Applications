```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv('House_prediction.csv')
```

```python
df.head()
```

```python
df['area'].mean()
```

```python
# Now we group the rows together by cities. The values corresponding to cities will be respective means.
df_mean=df.groupby(['city']).mean() 
```

```python
df_mean.head()
```

```python
for i in range(9):
   df_mean.plot.bar(y=df_mean.columns[i])
```

```python
from pandas.plotting import scatter_matrix
attributes=['area','rooms','bathroom','parking spaces',	'floor',	'animal','furniture','hoa (R$)',	'rent amount (R$)',	'property tax (R$)', 'fire insurance (R$)',	'total (R$)']
print(scatter_matrix(df[attributes], figsize=(20,15)))
```

```python
# The parameters vary heavily from one variable to another.
# Fire Insurance depends almost linearly on rent amount. 
# For rest of the features, either the values are concentrated near the y axis or assume various different values at the same x.
# property tax doesn't have an strong relationship with any variable. Same goes for HOA.
```
