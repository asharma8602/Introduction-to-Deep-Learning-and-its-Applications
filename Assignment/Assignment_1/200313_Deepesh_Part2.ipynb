```python
import numpy as np
import pandas as pd
```

```python
house= pd.read_csv('House_prediction.csv')
```

```python
house
```

```python
h=house.copy()
```

```python
h=h.groupby(['city']).mean()
```

```python
h=h.reset_index()
h
```

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

for y in h:
    if y != 'city':
        plt.plot(h['city'],h[y])
        plt.title(y)
        plt.ylabel(y)
        plt.show()
        print()
```
