```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\AMBUJ PYNE\Downloads\House_prediction.csv')

#print(df.to_string())
df1 = df[df['city'].str.contains('Belo Horizonte')]
df2 = df[df['city'].str.contains('Campinas')]
df3 = df[df['city'].str.contains('Porto Alegre')]
df4 = df[df['city'].str.contains('Rio de Janeiro')]
df5 = df[df['city'].str.contains('SÃ£o Paulo')]
print(df5)
```

```python
# Relation in Rent amont and Area
#df1.plot()
df1.plot(kind='scatter',x='area',y='hoa (R$)',title='Belo Horizonte',xlim=(0,500),ylim=(0,0.01)) # scatter plot
df2.plot(kind='scatter',x='area',y='hoa (R$)',color='g',title='Campinas',s=1,xlim=(0,400),ylim=(-10,8000))
df3.plot(kind='scatter',x='area',y='hoa (R$)',color='y',title='Porto Alegre',s=1,xlim=(0,400),ylim=(-10,4000))
df4.plot(kind='scatter',x='area',y='hoa (R$)',color='r',title='Rio de Janeiro',s=1,xlim=(0,1000),ylim=(-10,5000))
df5.plot(kind='scatter',x='area',y='hoa (R$)',color='b',title='SÃ£o Paulo')

plt.show()
```

```python
df1.plot(kind='scatter',x='rent amount (R$)',y='property tax (R$)',title='Belo Horizonte',xlim=(0,14000),ylim=(0,3000)) # scatter plot
df2.plot(kind='scatter',x='rent amount (R$)',y='property tax (R$)',color='g',title='Campinas')
df3.plot(kind='scatter',x='rent amount (R$)',y='property tax (R$)',color='y',title='Porto Alegre')
df4.plot(kind='scatter',x='rent amount (R$)',y='property tax (R$)',color='r',title='Rio de Janeiro',xlim=(0,14000),ylim=(0,5000))
df5.plot(kind='scatter',x='rent amount (R$)',y='property tax (R$)',color='b',title='SÃ£o Paulo')
```
