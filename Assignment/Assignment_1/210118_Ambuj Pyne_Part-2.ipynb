```python
import pandas as pd

df = pd.read_csv(r'C:\Users\AMBUJ PYNE\Downloads\House_prediction.csv')

#print(df.to_string())
df1 = df[df['city'].str.contains('Belo Horizonte')]
df2 = df[df['city'].str.contains('Campinas')]
df3 = df[df['city'].str.contains('Porto Alegre')]
df4 = df[df['city'].str.contains('Rio de Janeiro')]
df5 = df[df['city'].str.contains('SÃ£o Paulo')]

print(df1)
```

```python
print(df1[["area","rooms","bathroom","parking spaces","hoa (R$)",
          "rent amount (R$)","property tax (R$)","fire insurance (R$)","total (R$)"]].mean(),
          df2[["area","rooms","bathroom","parking spaces","hoa (R$)",
          "rent amount (R$)","property tax (R$)","fire insurance (R$)","total (R$)"]].mean())
```

```python
#AREA
import matplotlib.pyplot as plt
city = ['Belo Horizonte','Campinas','Porto Alegre','Rio de Janeiro','SÃ£o Paulo']
m_area = [df1["area"].mean(),df2["area"].mean(),df3["area"].mean(),df4["area"].mean(),df5["area"].mean()]
  
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(city, m_area, color ='maroon',
        width = 0.4)
 
plt.xlabel("Cities")
plt.ylabel("Area")
plt.title("Mean Area Citywise plot")
plt.show()
```
