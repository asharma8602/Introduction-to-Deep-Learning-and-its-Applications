```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('House_prediction.csv')
print(df)
```

```python
df2 = df.set_index('city')
df2 = df2.replace("-",0)
df2['floor'] = df2['floor'].astype(int)
#df_sp = df2.loc["São Paulo",["rooms","area","floor","parking spaces","property tax (R$)","bathroom","rent amount (R$)","fire insurance (R$)","property tax (R$)","total (R$)"]]
df2_sp = df2.loc["São Paulo",~df2.columns.isin(['animal', 'furniture','rooms','floor','bathroom','parking spaces'])]
df2_sp_rem = df2.loc["São Paulo",["rooms","floor","parking spaces","bathroom"]]
df2_sp.mean(numeric_only=True).plot(kind ='bar',ylim=(0,6500), title = "São Paulo")
```

```python
df2_sp_rem.mean(numeric_only=True).plot(kind ='bar',ylim =(0,10),title = "São Paulo")
```
