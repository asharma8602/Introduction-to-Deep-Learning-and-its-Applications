```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv("/content/drive/MyDrive/Stamatics/House_prediction.csv")
grouped_data = df.groupby(["city"])[list(df.head(0))[2:]].mean()
grouped_data
```

```python
for i in range(grouped_data.shape[1]):
    plt.figure(figsize=(16,9))
    plt.plot(grouped_data.index.tolist(),grouped_data.loc[:,list(grouped_data.head(0))[i]])
    plt.ylabel(list(grouped_data.head(0))[i])
    plt.show()
```

```python
ax=sns.heatmap(grouped_data.corr(), annot=True)
```
