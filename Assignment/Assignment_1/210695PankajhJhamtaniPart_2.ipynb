#**Assignment Part 3**

For this part of the assignment to find the dependance of HoA, property tax and fire insurance on other features I will use two methods:


1.   Finding depndance on numerical features : area,rooms, bathrooms, parking spaces and floor; can be done by plotting dependent feature along with the independent feature on the same graph, city wise. This will show us trends of the dependent feature wrt the independent one. 
2.   Finding dependeance on features like pet animals accepted and furnishing of the house can be done by grouping data according to these features and then plotting a bar graph of the means with respect to these features



```python
for i in range(5) :
  df5=df3.iloc[:,5]
  df6=df3.iloc[:,i]
  plt.rcParams["figure.figsize"] = [7.00, 3.50]
  plt.rcParams["figure.autolayout"] = True
  fig, ax1 = plt.subplots()
  ax1.plot(df5,color='blue',label='HOA')
  ax2=ax1.twinx()
  ax2.plot(df6,color='green',label=df6.name[0])
  fig.tight_layout()
  ax1.legend(loc='lower left')
  ax2.legend(loc='lower right')
  plt.title(df5.name[0]+" vs "+df6.name[0])
  plt.show()
```

From the graph we can see that Hoa has in general direct relationship with the features. We can also see that it becomes an inverse relationship in Rio de janerio and sao Paalo sometimes

```python
for i in range(5) :
  df5=df3.iloc[:,7]
  df6=df3.iloc[:,i]
  plt.rcParams["figure.figsize"] = [7.00, 3.50]
  plt.rcParams["figure.autolayout"] = True
  fig, ax1 = plt.subplots()
  ax1.plot(df5,color='blue',label='Prop tax')
  ax2=ax1.twinx()
  ax2.plot(df6,color='green',label=df6.name[0])
  fig.tight_layout()
  ax1.legend(loc='lower left')
  ax2.legend(loc='lower right')
  plt.title(df5.name[0]+" vs "+df6.name[0])
  plt.show()
```

Dependance of property tax is in general direct with the features with very few exceptions

```python
for i in range(5) :
  df5=df3.iloc[:,8]
  df6=df3.iloc[:,i]
  plt.rcParams["figure.figsize"] = [7.00, 3.50]
  plt.rcParams["figure.autolayout"] = True
  fig, ax1 = plt.subplots()
  ax1.plot(df5,color='blue',label='Fire insur')
  ax2=ax1.twinx()
  ax2.plot(df6,color='green',label=df6.name[0])
  fig.tight_layout()
  ax1.legend(loc='lower left')
  ax2.legend(loc='lower right')
  plt.title(df5.name[0]+" vs "+df6.name[0])
  plt.show()
```

```python
df7=df.groupby(['animal','furniture'])
```

```python
df8=df7.agg(['mean'])
```

```python
iter=[5,7,8]
for i in iter:
  df9=df8.iloc[0:,i]
  toplot=list(df9)
  fig=plt.figure()
  sp=fig.add_subplot(111)
  x=[0,1.5,3,4.5]
  myxticks=['accept|furnished','accept|not furnished','not accept|furnished','not accept|not furnished']
  plt.xticks(x,myxticks)
  sp.bar(x,toplot)
  plt.title(df9.name[0])
  plt.show()
```

Hoa, Property Tax and fire insuration do not seem to have a simple increasing or decreasing relationship the facts like that animal is accepted or not or the house is furnished or not.
These independent features have some complicated effect on the dependent features
