```python
import pandas as pd
way= '/content/drive/MyDrive/Colab Notebooks/House_prediction.csv'
df = pd.read_csv(way)

import seaborn as sns
import matplotlib.pyplot as plt
g_data= df.groupby(["city"])[list(df.head(0))[2:]].mean()
g_data
for j in range(g_data.shape[1]):
    plt.figure(figsize=(16, 9))
    plt.plot(g_data.index.tolist(),g_data.loc[:,list(g_data.head(0))[j]])
    plt.ylabel(list(g_data.head(0))[j])
    plt.show()
```

```python
sns.set(rc = {'figure.figsize':(16,9)})
gjg = sns.heatmap(g_data.corr(), annot=True)
```
