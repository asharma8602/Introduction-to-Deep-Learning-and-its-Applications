```python
import pandas as pd
path= '/content/drive/MyDrive/House_prediction.csv'
df = pd.read_csv(path)

import seaborn as sns
import matplotlib.pyplot as plt
group_data= df.groupby(["city"])[list(df.head(0))[2:]].mean()
group_data
for i in range(group_data.shape[1]):
    plt.figure(figsize=(15, 10))
    plt.plot(group_data.index.tolist(),group_data.loc[:,list(group_data.head(0))[i]])
    plt.ylabel(list(group_data.head(0))[i])
    plt.show()
```

```python
sns.set(rc = {'figure.figsize':(15,10)})
heat_mp = sns.heatmap(group_data.corr(), annot=True)
```
