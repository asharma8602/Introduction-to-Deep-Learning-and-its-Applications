```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Stamatics /House_prediction.csv')
# print (df.head(2))
df.loc[df['floor']=='-', "floor"]=0
df.loc[df['furniture']=='furnished', "furniture"]=1
df.loc[df['furniture']=='not furnished', "furniture"]=0
df['floor'] = df['floor'].astype(int)
df['furniture']= df['furniture'].astype(int)
# print (df.head(2))
```
