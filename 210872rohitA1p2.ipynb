```python
plt.plot(df1.iloc[:,6],df1.iloc[:,1],color='b')
plt.plot(df1.iloc[:,6],np.log(df1.iloc[:,0]))
plt.plot(df1.iloc[:,6],df1.iloc[:,2]*2)
plt.plot(df1.iloc[:,6],df1.iloc[:,3],color='k')
plt.plot(df1.iloc[:,6],np.log(df1.iloc[:,5]),color='red')
plt.plot(df1.iloc[:,6],np.log(df1.iloc[:,8]))
plt.show()
```

```python
plt.plot(df1.iloc[:,4],df1.iloc[:,1],color='b', label='ROOMS')
plt.plot(df1.iloc[:,4],np.log(df1.iloc[:,0])*0.5,color='r', label='AREA')
plt.plot(df1.iloc[:,4],df1.iloc[:,2]*2,color='g', label='BATHROOM')
plt.plot(df1.iloc[:,4],df1.iloc[:,3],color='k', label='PARKING SPACES')
plt.plot(df1.iloc[:,4],np.log(df1.iloc[:,5]),color='red', label='RENT AMOUNT')
plt.plot(df1.iloc[:,4],np.log(df1.iloc[:,8]),color='b', label='TOTAL')
plt.show()
```

```python
plt.plot(df1.iloc[:,7],pow(15,df1.iloc[:,1]),color='b', label='ROOMS')
plt.plot(df1.iloc[:,7],(df1.iloc[:,0]*35),color='r', label='AREA')
plt.plot(df1.iloc[:,7],pow(25,df1.iloc[:,2]),color='g', label='BATHROOM')
plt.plot(df1.iloc[:,7],df1.iloc[:,3],color='k', label='PARKING SPACES')
plt.plot(df1.iloc[:,7],(df1.iloc[:,5]),color='red', label='RENT AMOUNT')
plt.plot(df1.iloc[:,7],np.log(df1.iloc[:,8]),color='b', label='TOTAL')

plt.show()
```
